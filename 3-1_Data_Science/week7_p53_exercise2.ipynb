```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
age = np.array([30, 40, 50, 60, 40])
income = np.array([200, 300, 500, 600, 300])
worked = np.array([10, 20, 20, 20, 20])
vacation = np.array([4, 4, 1, 2, 5])

data = np.array([age, income, worked, vacation]).astype('int64')
np.set_printoptions(precision=2, suppress=True)

print("Data array")
data
```

```python
# Population covariance matrix
popCovMatrix = np.cov(data, bias=True)

print("Population covariance matrix")
popCovMatrix
```

```python
sns.heatmap(popCovMatrix, annot=True, fmt='g')
plt.show()
```

```python
# Sample covariance matrix
sampleCovMatrix = np.cov(data, bias=False)

print("Sample covariance matrix")
sampleCovMatrix
```

```python
sns.heatmap(sampleCovMatrix, annot=True, fmt='g')
plt.show()
```
